##1. feladat: Páros számok kiiratása a [20; 46] intervallumon

```
def print_even_numbers(a = 20, b = 46):
  """
  Prints the even numbers on the [a, b] open interval
  """
  for i in range(a, b+1, 2):
    print(i)
  return

print_even_numbers()
```

##2. feladat: Prím számok kiírása 30-ig

```
def is_prime(x):
  """
  Determines if the given number is prime or not 
  """
  if x==2 or x==3: return True
  if x<2 or x%2==0: return False
  
  import math
  midway = math.floor(math.sqrt(x)) + 1
  for i in range(3, midway, 2):
    if not(x%i): return False
  return True

def print_prime_numbers(a = 0, b = 30):
  """
  Prints the prime numbers on the [a, b] open interval
  """
  for i in range(a, b+1):
    if is_prime(i):
      print(i)
      
  return

print_prime_numbers()
```

##3. feladat: Bekért szám előjelének megállapítása

```
def sign_of_number():
  """
  Determines if the given number entered by the user is positive, negative or 0
  Returns 0 if the input is a number
  Returns 1 if the input is not a number
  """
  print("Give me a number")
  user_input = input()
  try: x = float(user_input)
  except ValueError:
    print("{} is not a number.".format(user_input))
    return 1
  else:
    if x>0: print("{} is a positive number.".format(x))
    elif x==0: print("0 is 0.")
    else: print("{} is a negative number.".format(x))
    return 0

sign_of_number()
```

##4. feladat: Tömb elemeinek átlaga

```
def mean_of_array(arr):
  """
  Calculates the mean of the array elements
  """
  sum = 0
  for x in arr: sum+=x
  return sum/len(arr)

mean_of_array([1, 2, 3, 9, -5])
```
