## 1. feladat: $f(x) = \begin{cases} 2 \text{, ha } x \in [-4; 0) \\ 2 \cdot \cos(x) \text{, ha } x \in [0,4] \end{cases}$ függvény ábrázolása:

```
import numpy as np
from matplotlib import pyplot as plt
from scipy import linalg as la
```

## 2. feladat: Mátrix invertálása

```
def invert_matrix():
  '''
  Inverts the matrix given by the user, and prints it on the console.
  returns:
    - 0 if no problem occured
    - 1 if the user made a typo
    - 2 if the matrix is not quadratic
    - 3 if the matrix is singular
  '''
  tmp = input('Number of rows: ')
  try: R = int(tmp)
  except ValueError:
    print('\'{}\' is not a valid number'. format(tmp))
    return 1
  tmp = input('Number of columns: ')
  try: C = int(tmp)
  except ValueError:
    print('\'{}\' is not a valid number'. format(tmp))
    return 1

  if (C != R):
    print('The matrix is not quadratic, therefore cannot be inverted.')
    return 2

  M = np.zeros((R, C))
  
  for i in range(R):
    for j in range(C):
      tmp = input('a_{}{} = '.format(i+1, j+1))
      try: M[i][j] = float(tmp)
      except ValueError:
        print('\'{}\' is not a valid number'. format(tmp))
        return 1

  D = np.linalg.det(M)

  if(not D):
    print('The matrix is singular, therefore cannot be inverted.')
    return 3

  Mi = np.linalg.inv(M)

  print('\nThe given matrix:\n{}\n\nThe inverted matrix:\n{}'.format(M, Mi))

  return 0
  
invert_matrix()
```

# 3. feladat: Függvény bemutatása (3D ábrázolás)

```
from matplotlib import cm

fig, ax = plt.subplots(subplot_kw={"projection": "3d"})

X = np.linspace(-1, 10, 100)
Y = np.linspace(-10, 1, 100)
X, Y = np.meshgrid(X, Y)
R = np.sqrt(X**2 + Y**2)
Z = np.sin(R)

surf = ax.plot_surface(X, Y, Z, cmap=cm.coolwarm,
                       linewidth=0, antialiased=False)

ax.set_zlim(-1.01, 1.01)

fig.colorbar(surf, shrink=0.5, aspect=5)

plt.show()
```
